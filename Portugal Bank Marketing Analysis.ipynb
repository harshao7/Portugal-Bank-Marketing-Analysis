#  Q4. Exponential smoothing using Holt-Winterâ€™s technique and Forecast onion price for Mumbai marke

```python
# Model with triple exponential smoothing
# applicable when trend and seasonality exists

from statsmodels.tsa.holtwinters import ExponentialSmoothing

Mar_On_Exp_Sm = ExponentialSmoothing(Mar_On_price, seasonal_periods = 12, trend = 'add', seasonal = 'add').fit()
Mar_On_Exp_Sm.summary()
```

# 

```python
# Residual given by the model

Mar_On_res1 = pd.DataFrame(Mar_On_Exp_Sm.resid)
Mar_On_res1.head()
```

```python
Mar_On_res1.plot()
plt.title('Line plot of Residuals')
```

# 

```python
# Squaring resiuals / errors

Mar_On_se1 = pow(Mar_On_res1, 2)
Mar_On_se1.head()
```

```python
# avearge / mean of Squaring residuals / errors

Mar_On_mse1 = (Mar_On_se1.sum()) / len(Mar_On_se1)
print(Mar_On_mse1)
```

```python
# Root of average / mean of Squaring residuals / errors

Mar_On_rmse1 = sqrt(Mar_On_mse1)
Mar_On_rmse1
```

# 

```python
# foresting / predicting

Mar_On_pred2 = Mar_On_Exp_Sm.forecast(steps = 19)
Mar_On_pred2
```

# 

```python
# Plot actual and forecast

plt.plot(Mar_On_price, color = 'm')
plt.plot(Mar_On_pred2, color = 'g')
plt.legend(['Actual', 'Forest - ExponentialSmoothing'], bbox_to_anchor = (1, 1), loc = 4)
plt.show()
```
